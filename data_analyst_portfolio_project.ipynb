```python
# Treating null values or replacing all values in the Electronics category
df.loc[df['Category']=='Electronics','Size']="Not Applicable"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in the Accessories category
df.loc[df['Category']=='Accessories','Size']="Free Size"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in the Footwear category
df.loc[df['Category']=='Footwear','Size']="Not Available"
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Treating null values or replacing all values in the Clothing category
df.loc[(df["Category"]=="Clothing") & (df["Size"].isnull()), "Size"] = df[df["Category"]=="Clothing"]["Size"].mode()[0]
```

### Review Rating

```python
df.head()
```

```python
# mean of review ratings
df['Review Rating'].mean()
```

```python
# median of review ratings
df['Review Rating'].median()
```

```python
# mode of review ratings
df['Review Rating'].mode()
```

```python
df.groupby("Item Purchased")['Review Rating'].mean()
```

```python
#filling null values of Review with the mean of each Item Purchased
df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df.head()
```

### Previous Purchases

```python
# Filling Previous Purchases with zero(0)
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
df.isnull().sum()
```

### Purchase Amount (USD)

```python
df.groupby("Item Purchased")['Purchase Amount (USD)'].mean()
```

```python
#filling null values of Purchase Amount (USD) with the mean of each Item Purchased
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')['Purchase Amount (USD)'].transform('mean'))
```

```python
from sqlalchemy import create_engine
import urllib.parse

# your server & database
server = r'localhost\SQLEXPRESS'
database = 'data_analysis_project'   # change if needed

# connection string
connection_string = urllib.parse.quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server};"
    f"DATABASE={database};"
    f"Trusted_Connection=yes;"
)

# create engine
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={connection_string}")

print("✅ Connected to SQL Server!")
```

```python
table_name = "customer_behaviour_analysis"  # SQL table name you want to create

df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database 'data_analysis_project'.")
```
